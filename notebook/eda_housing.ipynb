```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style("whitegrid")
```

```python
df = pd.read_csv("D:/internship/task2/dataset/cleaned_housing_data.csv")

df.head()
```

```python
df.describe() #Summary statistics such as mean, standard deviation, minimum and maximum values were generated.
```

```python
df.median(numeric_only=True) #Median values were calculated to understand the central tendency of numerical variables.
```

```python
df.hist(figsize=(15,10))

plt.tight_layout()

plt.show()
```

```python
plt.figure(figsize=(6,4))

sns.boxplot(y=df["house_price_index"])

plt.title("House Price Index")

plt.show()
```

```python
plt.figure(figsize=(6,4))

sns.boxplot(
    y=df["personal_disposable_income"]
)

plt.title("Personal Disposable Income")

plt.show()
```

```python
numeric_df = df.select_dtypes(include=np.number)

plt.figure(figsize=(10,8))

sns.heatmap(
    numeric_df.corr(),
    annot=True,
    cmap="coolwarm"
)

plt.title("Correlation Matrix")

plt.show()
```

```python
sns.pairplot(
    df[
        [
            "house_price_index",
            "personal_disposable_income",
            "real_house_price_index",
            "affordability_proxy"
        ]
    ]
)

plt.show()
```

```python
df.skew(numeric_only=True)
```

```python
plt.figure(figsize=(10,5))

sns.lineplot(
    x="year",
    y="house_price_index",
    data=df
)

plt.title("House Price Index Over Time")

plt.show()
```

```python
plt.figure(figsize=(8,5))

sns.scatterplot(
    x="personal_disposable_income",
    y="house_price_index",
    data=df
)

plt.show()
```
